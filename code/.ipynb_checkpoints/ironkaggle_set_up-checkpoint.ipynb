- To improve the prediction model we would need to normalize the target to be closer to a normal distribution. We will proceed to filter without outliers. 

```python
df["price"].describe().T
```

```python
df_price = df["price"]
df_price = df_price.reset_index()
```

```python
def outlier_slayer(data): # automatically removes outliers based on Q1, Q3
    for column in data.select_dtypes(include=[np.number]):
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data
```

```python
df_price = outlier_slayer(df_price)
```

```python
df_price.set_index("index" ,inplace=True)
```

```python
price_values = df_price["price"].values
```

```python
price_values = list(price_values)
price_values
```

```python
price_wo_out = df[df["price"].isin(price_values)]
```

```python
price_wo_out.head(5)
```

```python
sns.histplot(price_wo_out["price"], color=color, kde=True)
```

- From here we will use **df_wo_out** as the df to manipulate

<h3 style="color: #A4064D;">Correlation between variables | Heatmap Multicollinearity</h3>

```python
#Correlation between target and other variables

df.corrwith(price_wo_out["price"]).sort_values(ascending=False)
```

```python
#Correlation 

df_corr = round(price_wo_out.corr(), 2)
```

```python
# Correlation Matrix-Heatmap Plot
mask = np.zeros_like(df_corr)
mask[np.triu_indices_from(mask)] = True # optional, to hide repeat half of the matrix

f, ax = plt.subplots(figsize=(25, 15))
sns.set(font_scale=1.5) # increase font size

ax = sns.heatmap(df_corr, mask=mask, annot=True, annot_kws={"size": 12}, linewidths=.5, cmap="rocket", fmt=".2f", ax=ax) # round to 2 decimal places
ax.set_title("Dealing with Multicollinearity", fontsize=20) # add title
plt.show()
```
